### **3. Importing Transformers & Torch (this includes adding HF_TOKEN as part of the gated repo in Secrets tab)**

```python
import transformers
import torch

model_id = "meta-llama/Meta-Llama-3-8B"

pipeline = transformers.pipeline(
    "text-generation", model=model_id, model_kwargs={"torch_dtype": torch.bfloat16}, device_map="auto"
)
pipeline("Hey how are you doing today?")
```

```python
pipeline("I am the best AI in the world because")
```
